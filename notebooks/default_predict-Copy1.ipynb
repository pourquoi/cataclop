```python
#bb = bets.groupby('id').filter(lambda r: r['num'].nunique() == 1)
bb = bets.groupby('id').filter(lambda r: (r['target'] == 'pred_mlp_1').sum() == 0 and (r['target'] == 'pred_knn_10_1').sum() == 0  )

f = 'sub_category'

for s in bb[f].value_counts().index:
    print("---\n" + s + "\t{:.2f}\n---".format(bb[bb[f] == s]['profit'].sum()))
    for t in bb['target'].value_counts().index:
        bbb = bb[(bb['target'] == t) & (bb['odds_ref'] < 20) & (bb[f] == s)]
        print("{:10s}: \t {:+.2f} \t {:+.0f} \t {:+.2f}".format(t, bbb.profit.sum(), bbb.profit.count(), bbb.profit.mean()) )
```
