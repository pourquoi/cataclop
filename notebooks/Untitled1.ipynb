```python
import pandas as pd
import numpy as np

import matplotlib
%matplotlib inline
```

```python
from cataclop.core import models
```

```python
players = models.Player.objects.filter(is_racing=True, odds__is_final=False, odds__is_final_ref=False).distinct()
```

```python
print(models.Player.objects.filter(is_racing=True, odds__is_final=False, odds__is_final_ref=False).distinct().query)
```

```python
len(players)
```

```python
start = players[0].race.start_at

for player in players.iterator():
    print("PLAYER {} {:d} {}".format(player.id, player.num, player.position))
    for odds in player.odds_set.all().order_by('-date'):
        print("{:d} {:d} {} : {:f}".format(odds.id, odds.player.id, start - odds.date, odds.value))
```
